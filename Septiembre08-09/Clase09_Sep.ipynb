```python
import pandas as pd
from IPython.display import Image
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from pandas.core.api import DataFrame
#Data taken from https://www.kaggle.com/datasets/goyalshalini93/car-data
df=pd.read_csv("CarPrice_Assignment.csv")
df.head()
```

```python
df['CarName'] = df['CarName'].str.split().str[0]
```

```python
#Separate Numerical and  Categorical data
numerical_columns = df.select_dtypes(include=['number'])
categorical_columns = df.select_dtypes(exclude=['number'])
```

```python
#Analiza la dependencia del valor del auto como función de cada variable categorica
import seaborn as sns

for categorical_column in categorical_columns.columns:
    sns.violinplot(data=df, x=categorical_column,  y=numerical_columns["price"], palette='plasma')
    #sns.barplot(x=categorical_column, y=numerical_columns["price"], data=df)
    plt.title(f'{"Valor del auto"} vs. {categorical_column}')
    plt.xlabel(categorical_column)
    plt.ylabel("Valor del auto")
    #plt.xticks(rotation=90)  # Rotate x-axis labels if necessary
    plt.show()
```

```python
#Analiza la dependencia del valor del auto como función de cada variable numerica
for numerical_column in numerical_columns.columns:
    plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
    sns.jointplot(x=numerical_column, y=numerical_columns["price"], data=df, kind='scatter')
    plt.title(f'{"Valor del auto"} vs. {categorical_column}')
    plt.xlabel(numerical_column)
    plt.ylabel("Valor del auto")
    #plt.xticks(rotation=90)  # Rotate x-axis labels if necessary
    plt.show()
```
